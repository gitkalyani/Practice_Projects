#### Saperating features and label



```python
x= df.drop("W",axis=1)
y= df["W"]
```

#### Feature scaling using standard scalarization


```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = pd.DataFrame(scaler.fit_transform(x), columns = x.columns)
x
```

### Model Building

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
```

### Linear Regression

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25,random_state=384)
```

```python
from sklearn.linear_model import LinearRegression


regression = LinearRegression()

regression.fit(x_train, y_train) 
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.metrics import mean_absolute_percentage_error as mape
    
y_pred = regression.predict(x_train)
print('Training Error : ', mape(y_train, y_pred))
  
y_pred = regression.predict(x_test)
print('Validation Error : ', mape(y_test, y_pred))
print()
```

#### Predict the wins from given features

```python
df.tail(2)
```

```python
print('No of predicted wins : ',regression.predict(scaler.transform([[650,5457,1324,260,36,148,426,1327,82,731,655,4.09,1,6,41,92]])))
```

### Random Forest Regressor

```python
x_train,x_test,y_transformed,y_test = train_test_split(x,y,test_size = 0.25, random_state = 241)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression
X, y = make_regression(n_features=16, n_informative=2, random_state=0, shuffle=False)
rfr = RandomForestRegressor(max_depth=6)
rfr.fit(X, y)
```

```python
df.tail(2)
```

```python
print(rfr.predict([[650,5457,1324,260,36,148,426,1327,82,731,655,4.09,1,6,41,92]]))
```

```python
RM = RandomForestRegressor()


RM.fit(x_train, y_train) 
y_pred = RM.predict(x_train)
print('Training Error : ', mape(y_train, y_pred))
  
y_pred = RM.predict(x_test)
print('Validation Error : ', mape(y_test, y_pred))
print()
```

### Predict the wins from given features

```python
print('No of wins : ',RM.predict(scaler.transform([[650,5457,1324,260,36,148,426,1327,82,731,655,4.09,1,6,41,92]])))
```

#### Let's check how well model fits the train data/ how could model learned

```python
# Adjusted R2 score

RM.score(x_train,y_train)
```
